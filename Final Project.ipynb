### Finding Missing Data

```python
n_missing = df.isnull().sum().sort_values(ascending=False)
percent_missing = df.isnull().sum().sort_values(ascending=False) / len(df)
n_missing_df = pd.DataFrame(data=[n_missing, percent_missing])
n_missing_df
```

The first row refers to how many missing values there are, and the second row refers to the percentage of missing values there are.
